```python
year =[2010 , 2011 , 2012 , 2013 , 2014 , 2015]
growth = [ 0.895, 0.91 , 0.919, 0.945 , 0.443 ,0.785]
plt.plot(year,growth)
```

```python
#axis label
plt.xlabel("year")
plt.ylabel("growth")
plt.legend()
plt.title("yield")
year =[2010 , 2011 , 2012 , 2013 , 2014 , 2015]
growth = [ 0.895, 0.91 , 0.919, 0.945 , 0.443 ,0.785]
plt.plot(year,growth)
```

```python
#plotting multiple lines 
year =[2010 , 2011 , 2012 , 2013 , 2014 , 2015]
apples = [ 0.895, 0.91 , 0.919, 0.945 , 0.443 ,0.785]
oranges= [ 0.95, 0.891 , 0.95, 0.45 , 0.43 ,0.85]
plt.plot(year,apples)
plt.plot(year,oranges)
plt.legend(["apples","oranges"])
```

```python
#line marker
plt.plot(year,apples,oranges,marker="o" , c ="r", ls="-")
plt.plot(year,apples,oranges,marker="x", c ="k", ls="--")
plt.legend(["apples","oranges"])
```
